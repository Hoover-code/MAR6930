```python
# import packages and functions
import pandas as pd
import matplotlib as plt
import numpy as np
from plotnine import ggplot,geom_boxplot, aes, scale_y_continuous, scales, stat_summary, labs
from mizani.labels import label_comma
# data_in = xl("A1:B103", headers=True)... use this to read in Python for Excel
```

```python
# Read in the data from an Excel file
data_in = pd.read_excel('~/Downloads/Data_in.xlsx')
data_in
```

```python
# Arrange the categories in 
# the education column in the following order: 'Masters Grad', 'College Grad', 'HS Grad'
e_categories = ['Masters Grad', 'College Grad', 'HS Grad']
data_in['Education'] = pd.Categorical(data_in['Education'], categories = e_categories, ordered = True)
```

```python
# Create a boxplot of income by education level
# Add a point to represent the mean income for each education level
# Use the label_comma() function to format the y-axis labels
(ggplot(data_in)+
    geom_boxplot(mapping = aes("Education", "Income"), varwidth = True)+
    scale_y_continuous(labels = label_comma())+
    stat_summary(mapping=aes(x="Education", y="Income"), fun_y=np.mean, geom="point", shape='x', size=4)+
    labs(title = "Income by Education Level")
)
```
